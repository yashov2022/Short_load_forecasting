```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
file_path = '/content/Copy of Active Power Load - 33_11KV - Godishala Substation .xlsx'
df = pd.read_excel(file_path)
```

```python
print(df.head())
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
def get_load_data(dt):
    row = df[df['datetime'] == dt]
    if not row.empty:
        return row['POWER (KW)'].values[0]
    else:
        return None
```

```python
def get_previous_hour_data(dt):
    previous_hour = dt - timedelta(hours=1)
    return get_load_data(previous_hour)
```

```python
def get_previous_day_data(dt):
    previous_day = dt - timedelta(days=1)
    return get_load_data(previous_day)
```

```python
current_datetime = datetime(2021,11,11, 14)
current_load = get_load_data(current_datetime)
previous_hour_load = get_previous_hour_data(current_datetime)
previous_day_load = get_previous_day_data(current_datetime)
```

```python
print(f"Current load at {current_datetime}: {current_load}")
print(f"Load at previous hour ({current_datetime - timedelta(hours=1)}): {previous_hour_load}")
print(f"Load at same hour on previous day ({current_datetime - timedelta(days=1)}): {previous_day_load}")
```

```python
import pandas as pd
#code for finding n hours
# Load the Excel file into a pandas dataframe
df = pd.read_excel('/content/Copy of Active Power Load - 33_11KV - Godishala Substation .xlsx')
df['datetime'] = pd.to_datetime(df['datetime'])
df.set_index('datetime', inplace=True)
# Define the function to extract previous days' power data
def get_previous_days_power(hour, date):
    # Filter the data for the specified hour
    hour_data = df[df.index.hour == hour]

    # Filter the data for all days before the specified date
    previous_days_data = hour_data[hour_data.index.date < date.date()]

    # Return the power generated values for the previous days
    return previous_days_data['POWER (KW)']

# Example usage:
hour = 1  # specify the hour
date = pd.Timestamp('2021-11-12 01:00:00')  # specify the date (already in datetime format)

previous_days_power = get_previous_days_power(hour, date)

print(previous_days_power)
```
